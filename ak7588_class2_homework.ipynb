# 1. Solution:

```python
import requests
import pandas as pd

url = 'http://data.cityofnewyork.us/api/views/jb7j-dtam/rows.json'
results = requests.get(url).json() 
# results.keys()
# results['data']
df = pd.DataFrame(results['data'])
```

```python
# adding columns
columns = results['meta']['view']['columns']
columns[0]
```

```python
headers = [i['fieldName'] for i in columns]
df = pd.DataFrame(results['data'], columns = headers)
df.head()
```

```python
# run this once only
df.drop(labels = [':updated_at', ':created_at', ':id', ':sid', ':position', ':meta', ':created_meta', ':updated_meta'], axis=1, inplace=True)
```

```python
df.drop(df.index[1094:], inplace=True)
df.tail()
```

```python
# renaming columns
renaming = {'year': 'Year', 'leading_cause': 'Leading cause', 'sex': 'Sex', 'race_ethnicity': 'Race/Ethnicity', 'deaths': 'Deaths', 'death_rate': 'Death rate', 'age_adjusted_death_rate': 'Age adjusted death rate'	} # renaming_dict
df.rename(columns = renaming, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df["Year"] = pd.to_numeric(df['Year'])
df.dtypes
```

```python
df['Deaths'] = pd.to_numeric(df['Deaths'], errors = 'coerce')
df['Death rate'] = pd.to_numeric(df['Death rate'], errors = 'coerce')
df['Age adjusted death rate'] = pd.to_numeric(df['Age adjusted death rate'], errors = 'coerce')
df.dtypes
```

```python
df["Sex"] = pd.Categorical(df['Sex']) # categorical variables
df["Leading cause"] = pd.Categorical(df['Leading cause'])
df["Race/Ethnicity"] = pd.Categorical(df['Race/Ethnicity'])

df.head()
```
